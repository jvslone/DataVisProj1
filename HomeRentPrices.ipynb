## Zillow Data (Purchase Price) (2000-2025/1mo) (Low/Mid/High)

https://www.zillow.com/research/data/

```python
#Load the datasets
TopHomeData = pd.read_csv('Data/Housing/Zillow/TopTierHomeData.csv')
MidHomeData = pd.read_csv('Data/Housing/Zillow/MidTierHomeData.csv')
BottomHomeData = pd.read_csv('Data/Housing/Zillow/BottomTierHomeData.csv')

#Display the first few rows of each dataset
display(TopHomeData.head())
display(MidHomeData.head())
display(BottomHomeData.head())
```

```python
#Reformat to just date and price
def CreateDatePriceFrame(df):
    df = df[df['RegionName'] == 'United States']
    df = df.drop(columns=['RegionID', 'SizeRank','RegionType','StateName'])
    df = df.melt(id_vars=['RegionName'], var_name='Date', value_name='Price')
    df = df.drop(columns=['RegionName'])
    return df

#Create reformatted dataframes
TopHomeDataDF = CreateDatePriceFrame(TopHomeData)
MidHomeDF = CreateDatePriceFrame(MidHomeData)
BottomHomeDF = CreateDatePriceFrame(BottomHomeData)

#Display the first few rows of each reformatted dataframe
display(TopHomeDataDF.head())
display(MidHomeDF.head())
display(BottomHomeDF.head())
```

```python
#Plot the data
fig = plt.figure(figsize=(12, 6))
TD = [dt.datetime.strptime(d, '%Y-%m-%d') for d in TopHomeDataDF['Date']]
MD = [dt.datetime.strptime(d, '%Y-%m-%d') for d in MidHomeDF['Date']]
BD = [dt.datetime.strptime(d, '%Y-%m-%d') for d in BottomHomeDF['Date']]
plt.plot(TD, TopHomeDataDF['Price'], label='Top Tier Homes', color='green')
plt.plot(MD, MidHomeDF['Price'], label='Mid Tier Homes', color='orange')
plt.plot(BD, BottomHomeDF['Price'], label='Bottom Tier Homes', color='red')
plt.xlabel('Date')
plt.ylabel('Median Home Price (USD)')
plt.title('US Housing Prices by Tier (2000-2025)')
plt.grid()
dates_ticks = np.array(['2000-01-01','2005-01-01','2010-01-01','2015-01-01','2020-01-01','2025-01-01'])
dates_ticksl = [dt.datetime.strptime(d, '%Y-%m-%d') for d in dates_ticks]
plt.xticks(dates_ticksl)
plt.legend()
```

## Zillow Data (Rental Price) (2015-2025) (Mid)

https://www.zillow.com/research/data/

```python
#Load the dataset
MidRentData = pd.read_csv('Data/Housing/Zillow/MidTierRentData.csv')

#Display the first few rows of the dataset
display(MidRentData.head())
```

```python
#Create reformatted dataframe
MidRentDF = CreateDatePriceFrame(MidRentData)

#Display the first few rows of the reformatted dataframe
display(MidRentDF.head())
```

```python
#Plot the data
fig = plt.figure(figsize=(12, 6))
MRD = [dt.datetime.strptime(d, '%Y-%m-%d') for d in MidRentDF['Date']]
plt.plot(MRD, MidRentDF['Price'], label='Mid Tier Rent', color='blue')
plt.xlabel('Date')
plt.ylabel('Median Rental Price (USD/month)')
plt.title('US Rental Prices (2015-2025)')
plt.grid()
dates_ticks = np.array(['2015-01-01','2017-07-01','2020-01-01','2022-07-01','2025-01-01'])
dates_ticksl = [dt.datetime.strptime(d, '%Y-%m-%d') for d in dates_ticks]
plt.xticks(dates_ticksl)
plt.legend()
```

## Census Home Data (Unadjusted) (1940-2000/10yr) (Median)

https://www.census.gov/data/tables/time-series/dec/coh-values.html

```python
#Load the dataset
CensusHomeData = pd.read_csv('Data/Housing/CensusHomeData.csv')

#Display the first few rows of the dataset
display(CensusHomeData.head())
```

```python
#Reformat to just date and price
def CreateDatePriceFrame(df):
    df = df[df['State'] == 'United States']
    df = df.melt(id_vars=['State'], var_name='Date', value_name='Price')
    df = df.drop(columns=['State'])
    return df

#Create reformatted dataframe
CensusHomeDF = CreateDatePriceFrame(CensusHomeData)

#Display the first few rows of the reformatted dataframe
display(CensusHomeDF.head())
```

```python
#Plot the data
fig = plt.figure(figsize=(12, 6))
MCD = [dt.datetime.strptime(d, '%Y') for d in CensusHomeDF['Date']]
plt.plot(MCD, CensusHomeDF['Price'], label='Mid Tier Home', color='blue',  marker='o')
plt.xlabel('Date')
plt.ylabel('Median Home Price (USD)')
plt.title('US Housing Prices (1940-2000)')
plt.grid()
plt.xticks(MCD)
plt.legend()
```

## FED Housing Data (1965-2025/3mo) (Median SFH)

https://fred.stlouisfed.org/series/MSPUS

```python
#Load the dataset
FedHomeData = pd.read_csv('Data/Housing/FEDSFHMedianData.csv')

#Display the first few rows of the dataset
display(FedHomeData.head())
```

```python
#Plot the data
fig = plt.figure(figsize=(12, 6))
MFD = [dt.datetime.strptime(d, '%Y-%m-%d') for d in FedHomeData['observation_date']]
plt.plot(MFD, FedHomeData['MSPUS'], label='Mid Tier Home', color='blue')
plt.xlabel('Date')
plt.ylabel('Median Home Price (USD)')
plt.title('US Median Single-Family Home Prices (1965-2025)')
plt.grid()
dates_ticks = np.array(['1965-01-01','1975-01-01','1985-01-01','1995-01-01','2005-01-01','2015-01-01','2025-01-01'])
dates_ticksl = [dt.datetime.strptime(d, '%Y-%m-%d') for d in dates_ticks]
plt.xticks(dates_ticksl)
plt.legend()
```
